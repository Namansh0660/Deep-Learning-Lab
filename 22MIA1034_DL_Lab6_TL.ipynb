## Lab_A_4 - Transfer Learning

# Loading Libraries

```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing import image
import numpy as np
```

# Step 1 - Choosing a pretrained model

```python
# (Step-1) : Choose a pretrained model
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

# Step 2 - Freezing the initial layers

```python
# (Step-2) : Freeze the initial layers
for layer in base_model.layers:
    layer.trainable = False
```

# Step 3 - Replacing the final layers

```python
# (Step-3) : Replace the final layers
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
output_layer = Dense(1, activation='sigmoid')(x)
```

# Retraining the model

```python
# Retraining the model
model = Model(inputs=base_model.input, outputs=output_layer)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy,precision,recall'])
```

# Image Preprocessing

```python
# Image preprocessing
img_path = '/content/tench.jpeg'
img = image.load_img(img_path, target_size=(224, 224))
img_array = np.array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array = tf.keras.applications.resnet50.preprocess_input(img_array)
```

# Prediction

```python
# Prediction
p = model.predict(img_array)
p_img = np.argmax(p, axis=1)
print(p_img)
print(p)
class_names = ['tench','goldfish','great white shark','tiger shark','hammerhead','electric ray','stingray','cock','hen','ostrich']
print(f"Predicted_Class:{class_names[p_img[0]]}")
```
