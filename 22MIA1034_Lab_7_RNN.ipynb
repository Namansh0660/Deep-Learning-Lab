```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from sklearn.model_selection import train_test_split
```

# Loading the Dataset

```python
df = pd.read_csv("/content/IMDB Dataset.csv")
print(df.head())
```

# Text to Tokens

```python
# convert text to tokens
t = Tokenizer(num_words=5000)
t.fit_on_texts(df['review'])
x = t.texts_to_sequences(df['review'])
max_length = 100
xp = pad_sequences(x, maxlen=max_length, padding="post")
y = np.array(df['sentiment'].map({'positive':1, 'negative':0}))
```

# Train Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(xp, y, random_state=42, test_size=0.2)
```

# Model Training

```python
model = Sequential([
    Embedding(input_dim = 5000, output_dim = 32, input_length = max_length),
    SimpleRNN(16, activation = 'relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

# Printing Loss and Accuracy

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Loss: {loss}, Accuracy: {accuracy}")
```

# Prediction

```python
st = ["This is an amazing movie"]
ss = t.texts_to_sequences(st)
ssq = pad_sequences(ss, maxlen=max_length, padding="post")
p = model.predict(ssq)
print("Predicted Sentiment : ", "Positive" if p[0][0]>0.5 else "Negative")
```
