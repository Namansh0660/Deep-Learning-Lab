# Importing Libraries

```python
import pandas as pd
import numpy as np
```

# Loading the Dataset

```python
df = pd.read_csv('/content/Dataset.csv')
df
```

```python
print(df.head())
```

```python
t = df['Text'].values
t
```

```python
l = df['Label'].values
l
```

# Feature Extraction

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer

v = CountVectorizer()
f = v.fit_transform(t)

v1 = TfidfVectorizer()
f1 = v1.fit_transform(t)

print(f)
print("\n")
print(f1)
```

# MLP

```python
from sklearn.neural_network import MLPClassifier

ml = MLPClassifier()
ml.fit(f1,l)
```

# Testing

```python
text_t = ['The movie is bad.']
f_t = v1.transform(text_t)
ml.predict(f_t)
```

### Implement the following algorithms
* SVM
* Naive Bayes
* Random Forest

## SVM

```python
from sklearn.svm import SVC
sv = SVC()
sv.fit(f1,l)
```

```python
svm_output = sv.predict(f_t)
svm_output
```

# Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(f1,l)
```

```python
nb_output = nb.predict(f_t)
nb_output
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(f1,l)
```

```python
rf_output = rf.predict(f_t)
rf_output
```
